```r
# Start with mixed values
num_val <- 123.45
char_val <- "456"
log_val <- TRUE


# Numeric conversions
as.integer(num_val)    # 123 (truncates decimal)
as.character(num_val)  # "123.45"
as.numeric(char_val)   # 456


# Character conversions
as.numeric(char_val)   # 456
as.integer("123")      # 123
as.logical("TRUE")     # TRUE


# Logical conversions
as.numeric(log_val)    # 1 (TRUE=1, FALSE=0)
as.character(log_val)  # "TRUE"
```

```r
cat("=== R ARITHMETIC OPERATORS DEMO ===\n\n")


# Define sample values
a <- 15
b <- 4
cat("Values: a =", a, ", b =", b, "\n\n")


# 1. Basic Arithmetic
cat("1. BASIC OPERATIONS:\n")
cat("Addition (+):     ", a + b, "\n")
cat("Subtraction (-):  ", a - b, "\n")
cat("Multiplication (*):", a * b, "\n")
cat("Division (/):     ", a / b, "\n")
cat("Power (^):        2^3 =", 2^3, "\n\n")
```

```r
# 2. Special Division Operators
cat("2. DIVISION OPERATORS:\n")
cat("Modulo (%%):      ", a %% b, " (remainder)\n")
cat("Integer Div (%/%):", a %/% b, " (drops decimal)\n")
cat("Normal Div (/):   ", a / b, " (3.75)\n\n")
```

```r
# 3. Vector Operations (Element-wise)
cat("3. VECTOR ARITHMETIC:\n")
x <- c(10, 20, 30)
y <- c(2, 4, 5)
cat("Vector x:", x, "\n")
cat("Vector y:", y, "\n")
cat("x + y:   ", x + y, "\n")
cat("x * y:   ", x * y, "\n")
cat("x / y:   ", x / y, "\n\n")
```

```r
# 4. Real-World Example: Rectangle Area
length <- 12.5
width <- 8.2
area <- length * width
cat("4. RECTANGLE AREA:\n")
cat("Length =", length, ", Width =", width, "\n")
cat("Area =", area, "\n\n")
```

```r
# 5. Image Processing Example (Pixel Math)
cat("5. IMAGE PIXEL EXAMPLE:\n")
r <- 255    # Red channel
g <- 128    # Green channel
b <- 64     # Blue channel
brightness <- (r + g + b) / 3
cat("RGB values: R=", r, "G=", g, "B=", b, "\n")
cat("Average brightness:", round(brightness), "\n")
```

```r
cat("=== R COMPARISON OPERATORS DEMO ===\n\n")


# Define sample values
a <- 25
b <- 18
c <- 25
cat("Values: a =", a, ", b =", b, ", c =", c, "\n\n")


# 1. Scalar Comparisons
cat("1. SCALAR COMPARISONS:\n")
cat("a == c (Equal):          ", a == c, "\n")
cat("a != b (Not Equal):      ", a != b, "\n")
cat("a > b (Greater):         ", a > b, "\n")
cat("b < a (Less):            ", b < a, "\n")
cat("a >= c (Greater/Equal):  ", a >= c, "\n")
cat("b <= a (Less/Equal):     ", b <= a, "\n\n")
```

```r
# 2. Vector Comparisons (Element-wise)
cat("2. VECTOR COMPARISONS:\n")
x <- c(10, 20, 30, 25)
y <- c(15, 20, 25, 30)
cat("Vector x:", x, "\n")
cat("Vector y:", y, "\n")
cat("x == y:    ", x == y, "\n")
cat("x > y:     ", x > y, "\n")
cat("x >= y:    ", x >= y, "\n")
cat("x <= y:    ", x <= y, "\n\n")
```

```r
# 3. String Comparisons
cat("3. CHARACTER COMPARISONS:\n")
place1 <- "Tirupati"
place2 <- "Tambaram"
place3 <- "Tirupati"
cat("place1 == place3: ", place1 == place3, "\n")
cat("place1 != place2: ", place1 != place2, "\n\n")
```

```r
# 4. Practical Example: Student Grade Filter
cat("4. STUDENT GRADES FILTER:\n")
scores <- c(85, 72, 95, 68, 88)
pass_threshold <- 75
failed <- scores < pass_threshold
cat("Scores:", scores, "\n")
cat("Pass threshold:", pass_threshold, "\n")
cat("Failed students (scores < 75):", failed, "\n")
cat("Number failed:", sum(failed), "\n\n")
```

```r
# 5. Image Thresholding Example (Vision Relevant)
cat("5. IMAGE THRESHOLDING:\n")
pixel_values <- c(50, 120, 200, 80, 250)
threshold <- 100
bright_pixels <- pixel_values > threshold
cat("Pixel values:", pixel_values, "\n")
cat("Bright pixels (>100):", bright_pixels, "\n")
cat("Count of bright pixels:", sum(bright_pixels), "\n")
```

```r
cat("=== R LOGICAL OPERATORS DEMO ===\n\n")


# 1. Scalar Logical Values
cat("1. SCALAR LOGICALS:\n")
p <- TRUE
q <- FALSE
cat("p =", p, ", q =", q, "\n")
cat("p & q (AND):     ", p & q, "\n")
cat("p && q (AND):    ", p && q, "\n")
cat("p | q (OR):      ", p | q, "\n")
cat("p || q (OR):     ", p || q, "\n")
cat("!p (NOT p):      ", !p, "\n")
cat("!q (NOT q):      ", !q, "\n\n")
```

```r
# 2. From Comparisons
cat("2. FROM COMPARISONS:\n")
a <- 25
b <- 18
c <- 30
cat("a =", a, ", b =", b, ", c =", c, "\n")
cat("(a > b) & (b < c):", (a > b) & (b < c), "\n")
cat("(a > b) && (b < c):", (a > b) && (b < c), "\n")
cat("(a < b) | (c > b):", (a < b) | (c > b), "\n\n")
```

```r
# 3. Vector Logical Operations (Element-wise)
cat("3. VECTOR LOGICALS (& and | are VECTORIZED):\n")
x <- c(TRUE, FALSE, TRUE, FALSE)
y <- c(FALSE, TRUE, TRUE, FALSE)
cat("Vector x:", x, "\n")
cat("Vector y:", y, "\n")
cat("x & y: ", x & y, "\n")
cat("x | y: ", x | y, "\n")
cat("!(x):  ", !x, "\n\n")
```

```r
# 4. Practical Example: Student Eligibility Filter
cat("4. STUDENT ELIGIBILITY:\n")
scores <- c(85, 72, 95, 68, 88)
attendance <- c(95, 82, 98, 60, 90)  # percent
min_score <- 75
min_attend <- 80
eligible <- (scores >= min_score) & (attendance >= min_attend)
cat("Scores:", scores, "\n")
cat("Attendance:", attendance, "\n")
cat("Eligible:", eligible, "\n")
cat("Eligible count:", sum(eligible), "out of", length(eligible), "\n\n")
```

```r
# 5. Image Processing: Multi-Condition Mask
cat("5. IMAGE MASKING (Vision Example):\n")
is_bright <- c(TRUE, FALSE, TRUE, FALSE, TRUE)  # brightness > threshold
is_edge <- c(FALSE, TRUE, TRUE, FALSE, TRUE)     # edge detected
valid_pixels <- is_bright & is_edge              # Both conditions
cat("Bright pixels:", is_bright, "\n")
cat("Edge pixels:  ", is_edge, "\n")
cat("Valid pixels: ", valid_pixels, "\n")
cat("Valid count:  ", sum(valid_pixels), "\n")
```

```r
marks <- c(85, 42, 78, 90, 66, 58, 73, 49, 91, 60)
grades <- c()

for (m in marks) {
  if (m >= 75) {
    grades <- c(grades, "Distinction")
  } else if (m >= 60) {
    grades <- c(grades, "First Class")
  } else if (m >= 50) {
    grades <- c(grades, "Second Class")
  } else {
    grades <- c(grades, "Fail")
  }
}

grades
```
