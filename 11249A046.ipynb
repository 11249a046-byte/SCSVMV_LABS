```r
# Simulated brightness levels (Vector in R)
frame_brightness <- c(45, 210, 120, 10, 255, 180, 5, 130, 90, 202)
processed_frames <- c()


cat("--- Image Quality Report ---\n")


# Using a FOR loop
for (i in 1:length(frame_brightness)) {
  brightness <- frame_brightness[i]

  # IF/ELSE logic
  if (brightness < 50) {
    status <- "Underexposed (Discard)"
  } else if (brightness > 200) {
    status <- "Overexposed (Discard)"
  } else {
    status <- "Optimal (Save)"
    processed_frames <- c(processed_frames, brightness)
  }

  cat(sprintf("Frame %d: Brightness %d -> %s\n", i, brightness, status))
}


cat("\nTotal usable frames for processing:", length(processed_frames), "\n")
```

```r
electricity_bill <- function(units){
if(units <= 100){
bill <- units * 2
} else if(units <= 200){
bill <- (100 * 2) + (units - 100) * 3
} else {
bill <- (100 * 2) + (100 * 3) + (units - 200) * 5
}
return(bill)
}
electricity_bill(250)
```

Question
06:

University Student Data Cleaning System

A university collects student data from multiple online forms.

Due to manual entry:

Names have extra spaces

Emails are in mixed case

Department codes are inconsistent (cse, CSE , Cse)

Email usernames need to be extracted

Only valid university emails (@univ.edu) should be retained

Raw Data (Before Cleaning)

```r
students <- data.frame(
Name = c(" Gayathri ", "RAVI ", " meena"),
Email = c("Gayu@UNIV.edu", "ravi@univ.edu", "meena@gmail.com"),
Department = c(" cse", "ECE ", " Cse ")
)
students
students$Name <- trimws(students$Name)
students$Department <- trimws(students$Department)
students$Name <- toupper(students$Name)
students$Email <- tolower(students$Email)
students$Department <- toupper(students$Department)
students$Username <- sub("@.*", "", students$Email)
students$ValidEmail <- grepl("@univ.edu$", students$Email)
clean_students <- students[students$ValidEmail == TRUE, ]
clean_students
```

9. Create a data set and do statistical analysis on the data using R

Question
1:

School Student Marks Analysis

Scenario

A teacher wants to analyze marks of students in a class. She collects marks for Maths, Science, and
English for 10 students.

Tasks:

1. Create a dataset in R.

2. Calculate mean, median, and standard deviation of marks for each subject.
3. Identify the student with highest total marks.
4. Find students scoring above average in Maths.

R Program

```r
# Step 1: Create dataset
students <- data.frame(
Name =
c("Arjun","Meena","Rahul","Anitha","Kiran","Sneha","Ravi","Gayathri","Suresh","Lakshmi"),
Maths = c(85,90,70,95,80,88,75,92,68,86),
Science = c(78,88,60,92,85,80,70,90,75,82),
English = c(92,80,75,98,88,85,78,91,80,87)
)

# Step 2: Mean, Median, SD
mean_marks <- sapply(students[ ,2:4], mean)
median_marks <- sapply(students[ ,2:4], median)
sd_marks <- sapply(students[ ,2:4], sd)
# Step 3: Total marks per student
students$Total <- rowSums(students[ ,2:4])
top_student <- students[which.max(students$Total), ]
# Step 4: Students above average in Maths
avg_math <- mean(students$Maths)
above_avg_math <- students[students$Maths > avg_math, ]
mean_marks
median_marks
sd_marks
top_student
above_avg_math
```

Project title: Air Quality Index Monitoring

```r
# ============================================
# Project title : Air Quality Index Monitoring
# ============================================

# Step 1: Load required libraries

library(ggplot2)   # For creating visualizations
library(dplyr)     # For data manipulation

# Step 2: Define the AQI dataset inline
aqi_data <- data.frame(
  StationID = c("S001","S002","S003","S004","S005",
"S006","S007","S008","S009","S010"),
  Date      = as.Date(c("2024-01-15","2024-01-16","2024-01-17",
"2024-01-18","2024-01-19","2024-01-20",
"2024-01-21","2024-01-22","2024-01-23",
"2024-01-24")),
  City      = c("Chennai","Mumbai","Delhi","Kolkata","Bangalore",
"Chennai","Mumbai","Delhi","Kolkata","Bangalore"),
  AQI       = c(142, 89, 210, 55, 78, 165, 38, 295, 110, 62),
  Pollutant_Conc = c(45, 30, 60, 20, 25, 50, 10, 80, 35, 22),
  Primary_Pollutant = c("PM2.5","PM10","PM2.5","CO","NO2",
"PM2.5","SO2","PM2.5","PM10","O3"),
  Category  = c("Unhealthy","Moderate","Very Unhealthy","Good",
"Moderate","Unhealthy","Good","Hazardous",
"Unhealthy for Sensitive","Moderate")
)

# Step 3: View the dataset
View(aqi_data)
cat("AQI Dataset Loaded Successfully\n")
print(summary(aqi_data))

# Step 4: Bar Chart – Average AQI by City
avg_aqi <- aqi_data %>%
  group_by(City) %>%
  summarise(Avg_AQI = mean(AQI))   # Compute mean AQI per city

ggplot(avg_aqi, aes(x = City, y = Avg_AQI, fill = City)) +
  geom_bar(stat = "identity") +    # Draw bars with actual values
  labs(title = "Average AQI by City",
       x = "City", y = "Average AQI") +
  theme_minimal() +
  theme(legend.position = "none")  # Remove legend (redundant with x-axis)

# Step 5: Histogram – AQI Value Distribution
ggplot(aqi_data, aes(x = AQI)) +
  geom_histogram(binwidth = 40, fill = "steelblue", color = "black") +
  labs(title = "Distribution of AQI Values",
       x = "AQI", y = "Frequency") +
  theme_minimal()

# Step 6: Box Plot – AQI spread per City
ggplot(aqi_data, aes(x = City, y = AQI, fill = City)) +
  geom_boxplot() +                 # Show median, IQR, and outliers
  labs(title = "AQI Box Plot by City",
       x = "City", y = "AQI") +
  theme_minimal() +
  theme(legend.position = "none")

# Step 7: Pie Chart – Proportion of AQI Categories
category_counts <- table(aqi_data$Category)   # Count each category
pie(category_counts,
    main = "AQI Category Distribution",
    col  = rainbow(length(category_counts)))
```
